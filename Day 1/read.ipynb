```python
#reading the text file in python 
file = open("../Assets/test.txt", encoding="utf-8").read()
print(file)
```

```python
#cleaning the text
#1. Chnage the text to the lower case bacaue the python is case sensitive
#2. Remove the punctuation
#3. Remove the stopwords
#4. Remove the numbers
text_lower = file.lower()
print(text_lower)
```

```python
#Remove Punctuation 
#In english language, punctuation marks are of 16 types they are ! " # $ % & ' ( ) * + , - . / : ; ? @ [ \ ] ^ _ ` { | } ~
#We can remove punctuation marks from the text by using the following code

import string
#this will give all punctuation marks
print(string.punctuation)

```

```python
#now we will need to remove this from the data set
cleaned_data = text_lower.translate(str.maketrans('', '', string.punctuation))
print(cleaned_data)
```

```python
#Tokenization
#Tokenization is the process of breaking a stream of text up into words, phrases, symbols, or other meaningful elements called tokens.
tokenized_word= cleaned_data.split()
print(tokenized_word)
```

```python
#remove the stopwords
#stop words are the words which are not useful for the analysis
stop_words = ["i", "me", "my", "myself", "we", "our", "ours", "ourselves", "you", "your", "yours", "yourself",
              "yourselves", "he", "him", "his", "himself", "she", "her", "hers", "herself", "it", "its", "itself",
              "they", "them", "their", "theirs", "themselves", "what", "which", "who", "whom", "this", "that", "these",
              "those", "am", "is", "are", "was", "were", "be", "been", "being", "have", "has", "had", "having", "do",
              "does", "did", "doing", "a", "an", "the", "and", "but", "if", "or", "because", "as", "until", "while",
              "of", "at", "by", "for", "with", "about", "against", "between", "into", "through", "during", "before",
              "after", "above", "below", "to", "from", "up", "down", "in", "out", "on", "off", "over", "under", "again",
              "further", "then", "once", "here", "there", "when", "where", "why", "how", "all", "any", "both", "each",
              "few", "more", "most", "other", "some", "such", "no", "nor", "not", "only", "own", "same", "so", "than",
              "too", "very", "s", "t", "can", "will", "just", "don", "should", "now"]

#remove the stop words from the text
final_words = []
for word in tokenized_word: #loop throught the tokenized words
    if word not in stop_words: #check if the word is not in the stop words list
        final_words.append(word) #append the word to the final_words list
print(final_words)
```

```python
#above code removed the stop words
#now we will work to use NLP to find the most common words in the reviews
emotion_list = []
with open('../Assets/emotions.txt', 'r') as file:
    for line in file:
        clear_line = line.replace('\n', '').replace(',', '').replace("'", '').strip() #strip removes the spaces, \n is for new line, replace is to remove the commas and apostrophes
        word,emotion = clear_line.split(':')
        #print("Word :" + word + " " + "Emotion :" + emotion)
        if word in final_words:
            emotion_list.append(emotion)

print(emotion_list)
```

```python
#now we will count the emotions and plot it using matplotlib
import matplotlib.pyplot as plt
from collections import Counter

#counting the emotions
counting = Counter(emotion_list)
print(counting)
```

```python
#plot it in a graph
fig = plt.figure(figsize=(20,10))
plt.plot(counting.keys(), counting.values())
plt.show()

#now with small subplots
fig, axs = plt.subplots()
axs.bar(counting.keys(), counting.values())
plt.xlabel('Emotion')
plt.ylabel('Count')
plt.show()
```
